```python

hash=[]
for i in range(len(df['hashtags'])):
  hash.append(df['hashtags'][i])
```

```python
hash
dif_hash=[]
for i in range(len(hash)):
  for j in range(len(dif_hash)):
    if(hash[i]==dif_hash[j]):
      break
  else:
    dif_hash.append(hash[i])
```
